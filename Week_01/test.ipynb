```python
import numpy as np
import matplotlib.pyplot as plt

# Simulierte Daten erzeugen
daten = np.random.normal(loc=0, scale=1, size=1000)

# Histogram plotten
plt.hist(daten, bins=30, edgecolor='black')
plt.title('Histogramm der simulierten Daten')
plt.xlabel('Wert')
plt.ylabel('HÃ¤ufigkeit')
plt.grid(True)
plt.show()
```
